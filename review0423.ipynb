```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rc('font',family='NanumBarunGothic')
#plt.rc('font',ramily='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
```

```python
filename='성씨ㆍ본관별_인구__시군구_20240423154924.csv'
df=pd.read_csv(filename)
df
```

```python
df.columns=['성씨','지역','인구']
df.head(2)

df=df[(df['성씨']!='계')&(df['지역']=='전국')]
#df=df.query("성씨!='계' and 지역=='전국'")
df.head()
```

```python
set(df['성씨'])
df['성씨'].unique()
```

```python
len(df['성씨'].unique())
df['성씨'].unique()
```

```python
total=len(set(name.split('(')[0]for name in df.성씨.unique()))
names=set(name.split('(')[0]for name in df.성씨.unique())
print(sorted(list(names)))
```

```python
df_all=df.copy()
for idx in df_all.index:
    fname=df_all.loc[idx,'성씨'].split('(')[0]
    df_all.loc[idx,'성씨']=fname
```

```python
df_all[df_all['성씨']=='조']
df_all_group=df_all.groupby('성씨').sum(numeric_only=True)
df_all_group
```
