```python
filename='성씨ㆍ본관별_인구__시군구_20240423192727.csv'
df=pd.read_csv(filename)
df
```

```python
df.columns=['성씨','지역','인구']

df=df[(df.성씨!='계')&(df.지역!='전국')]
df.query("성씨!='계' and 지역!='전국'")
df.head()
```

```python
df_goo=df[df.성씨.str.contains('구')]
df_goo
```

```python
family_name='박'
df_my=df[df.성씨.str.contains(family_name)]
df_my
```

```python
df_goo_group=df_goo.groupby('지역').sum(numeric_only=True)

df_goo_group=df_goo_group.sort_values(by='인구',ascending=False)

df_goo_group
```

```python
df_my_group=df_my.groupby('지역').sum(numeric_only=True)
df_my_group=df_my_group.sort_values(by='인구',ascending=False)
df_my_group
```

```python
df_goo_group.columns
```

```python
import matplotlib.pyplot as plt

x=df_goo_group.index
y=df_goo_group.인구

plt.bar(x,y)

plt.title('대한민국 "구씨"를 찾아라!',fontsize=15)
plt.xlabel('지역')
plt.ylabel('인구')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=df_goo_group.index
y=df_goo_group.인구

plt.figure(figsize=(15,5),facecolor='lavender')
plt.bar(x,y)

plt.title('대한민국 "구씨"를 찾아라!',pad=20,
          fontdict={'size':30,'color':'black','weight':'bold'})
plt.xlabel('지역',fontdict={'size':20,'color':'dimgray','weight':'bold'})
plt.ylabel('인구',fontdict={'size':20,'color':'dimgray','weight':'bold'},labelpad=20)
plt.xticks(rotation=45)
plt.grid(True,axis='y',alpha=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rc('font',family='NanumBarunGothic')

x=df_goo_group.index
y=df_goo_group.인구

plt.figure(figsize=(15,5),facecolor='lavender')
plt.plot(x,y)

plt.title('대한민국 "구씨"를 찾아라!',pad=20,
          fontdict={'size':30,'color':'black','weight':'bold'})
plt.xlabel('지역',fontdict={'size':20,'color':'dimgray','weight':'bold'})
plt.ylabel('인구',fontdict={'size':20,'color':'dimgray','weight':'bold'},labelpad=20)
plt.xticks(rotation=45)
plt.grid(True,axis='y',alpha=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=df_goo_group.index
y=df_goo_group.인구

plt.figure(figsize=(15,5),facecolor='lavender')
plt.bar(x,y)
plt.plot(x,y,'-or')

plt.title('대한민국 "구씨"를 찾아라!',pad=20,
          fontdict={'size':30,'color':'black','weight':'bold'})
plt.xlabel('지역',fontdict={'size':20,'color':'dimgray','weight':'bold'})
plt.ylabel('인구',fontdict={'size':20,'color':'dimgray','weight':'bold'},labelpad=20)
plt.xticks(rotation=45)
plt.grid(True,axis='y',alpha=0.5)
plt.savefig('family_name_goo.png')
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

font_path='/usr/share/fonts/truetype/nanum/NanumGothicBold.ttf'

data=dict(zip(df_goo_group.index,df_goo_group.인구))

wc=WordCloud(width=1000,height=600,
             background_color='black',
             font_path=font_path)
plt.imshow(wc.generate_from_frequencies(data))
plt.axis('off')
plt.show()

wc.to_file('./구씨_워드클라우드.png')
```
